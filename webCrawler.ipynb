```python
import requests
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
```

```python
###setting###
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
```

```python
place = "台北"
driver = webdriver.Chrome("/Users/ymc/Desktop/chromedriver")
driver.get("https://www.tripadvisor.com.tw")

###save current main url###
main_window = driver.current_window_handle

###activate hotel search modal###
hotelBtn = driver.find_elements_by_css_selector('a[href="/Hotels"]')
hotelBtn[0].click()
time.sleep(1)

###input place and find###
searchBar = driver.find_elements_by_class_name("input-text-input-ManagedTextInput__managedInput--2RESp")
searchBar[0].send_keys(place)
time.sleep(1)
searchBar[0].send_keys(Keys.RETURN)

###remove calendar modal###
driver.find_element_by_id("HEADING").click()
time.sleep(0.5)

totalHotelPage = 1

for i in range(totalHotelPage):
    ###get all hotel list###
    hotel = driver.find_elements_by_class_name("property_title")
    
    ###start surfing all hotel###
    for j in hotel:
        j.click()
        time.sleep(2)
        driver.switch_to.window(driver.window_handles[1])
        print(driver.find_element_by_id("HEADING").text)
        driver.close()
        driver.switch_to.window(main_window)
        time.sleep(1)
```
